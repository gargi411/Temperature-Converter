```python
def convert_temperature():
    print("Temperature Converter")
    print("1. Celsius to Fahrenheit & Kelvin")
    print("2. Fahrenheit to Celsius & Kelvin")
    print("3. Kelvin to Celsius & Fahrenheit")

    choice = int(input("Enter your choice (1/2/3): "))
    value = float(input("Enter temperature value: "))

    if choice == 1:
        c = value
        f = (c * 9/5) + 32
        k = c + 273.15
        print(f"\nCelsius: {c}°C")
        print(f"Fahrenheit: {f:.2f}°F")
        print(f"Kelvin: {k:.2f}K")

    elif choice == 2:
        f = value
        c = (f - 32) * 5/9
        k = c + 273.15
        print(f"\nFahrenheit: {f}°F")
        print(f"Celsius: {c:.2f}°C")
        print(f"Kelvin: {k:.2f}K")

    elif choice == 3:
        k = value
        c = k - 273.15
        f = (c * 9/5) + 32
        print(f"\nKelvin: {k}K")
        print(f"Celsius: {c:.2f}°C")
        print(f"Fahrenheit: {f:.2f}°F")

    else:
        print("Invalid choice!")

convert_temperature()
```
